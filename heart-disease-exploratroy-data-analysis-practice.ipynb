# Correlation & Heatmap

```python
df.head()
```

```python
##trying to use df.corr but restingECG is coming with error, use this code to exclude, values are in percentage deci
df_numeric = df.drop(columns=['RestingECG']).select_dtypes(include=[float, int])
correlation_matrix = df_numeric.corr()
print(correlation_matrix)
```

```python
sns.heatmap(correlation_matrix, annot = True, fmt='.2f',linewidth = 0.5)
```
